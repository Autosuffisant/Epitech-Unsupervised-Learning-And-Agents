### PART 1 : DATA DISTRIBUTION AND THE LAW OF LARGE NUMBERS

### 1. Propose a 2-dimensional random variable Z = (X, Y), with X and Y being two real (âˆˆ R), discrete or continuous random variables.

> These two variables should represent a quantities of your choice (e.g. the age of the individuals in a population, the color of the eyes of these individuals, ...). Compute the expected value of Z, that must be finite.

### Variable definition

$$ Z = (X, Y)$$

- $X$ is the age of the individuals in a population (integer between 0 and 100)
- $Y$ is the gross yearly income (integer between 0 and 500000)

So the expected value of Z is:

$$ E[Z] = (E[X], E[Y]) = (50, 250000) $$

```python
EXPECTED_Z = ((0 + 100) / 2, (0 + 500000) / 2)

print("Expected value of Z =", EXPECTED_Z)
```

### 2. Sample a number n of points from the law of Z and plot them in a 2 dimensional figure.

For a good distribution, n should be large, however for a proper visualization, n has been set to 1000.


```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
MAX_AGE = 100
MAX_INCOME = 250000

X_DISTRIBUTION = np.random.randint(0, MAX_AGE, N)
Y_DISTRIBUTION = np.random.randint(0, MAX_INCOME, N)

plt.scatter(X_DISTRIBUTION, Y_DISTRIBUTION)
plt.xlabel("X: Individual's age")
plt.ylabel("Y: Individual's yearly income")
plt.show()
```

### 3. For increasing values n, compute the empirical average of the first n samples and verify that it converges to the expected value, by plotting the euclidean distance between the empirical average and the expected value as a function of n.

```python
EMPIRICAL_Z = (np.mean(X_DISTRIBUTION), np.mean(Y_DISTRIBUTION))

print("Empirical value of Z: ", EMPIRICAL_Z)

# Euclidean distance function used for plotting of the distance
# between the empirical average and the expected value
def get_euclidean_distance(x, y):
    return np.sqrt(np.sum((np.array(x) - np.array(y)) ** 2))

euclidean_distances = []

for n in range(1, N):
    empirical_average = (np.mean(X_DISTRIBUTION[:n]), np.mean(Y_DISTRIBUTION[:n]))
    distance = get_euclidean_distance(empirical_average, EXPECTED_Z)
    euclidean_distances.append(distance)

plt.plot(euclidean_distances)
plt.xlabel("Number of samples")
plt.ylabel("Euclidean distance between\n empirical average and expected Z")
plt.show()
```

# Observation

The empirical average converges to the expected value as **n** increases.
